# Ajuste de Curvas e Análise de Regressão

## Tratamento Matemático

```python
import numpy as np
# Criando os vetores que correspondem as medições de corrente
# (Vetor x) e medições de potencial (Vetor y)
x = np.array([0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40])
y = np.array([7.1, 9.6, 16.9, 21.0, 25.4, 28.1, 35.7, 39.0])

# Obtendo os valores de 'a' e 'b' para a função afim que melhor ajusta
# os dados observados
mod_linear = np.polyfit(x, y, 1)

a = mod_linear[0]
b = mod_linear[1]
y_mod = a * x + b


def lmod(a, b, x):
    '''
    Retorna o valor de uma função linear do tipo y = ax + b, onde 'a' é o coeficiente angular,
    'b' é o coeficiente linear e 'x' é o ponto onde se deseja calcular a função.
    '''
    return a * x + b
```

## Representação Gráfica

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.figure(figsize = (7,5), dpi=180) #Ajusta tamanho e resolução da visualização da figura no editor
plt.plot(x, y, 'o', label = 'Pontos Experimentais', color = 'r') # Plota um gráfico de pontos, pois possui o parametro 'o' declarado
plt.xlabel(r'$\mathrm{I}$', family='sans-serif', weight = 'bold') #Ajusta a legenda do eixo x
plt.ylabel(r'$\mathrm{V }$           ', rotation = 0,  family='sans-serif', weight = 'bold') # Ajusta a legenda do eixo y
#plt.title('Ajuste da Reta de Ohm', fontsize = 14)
plt.legend(frameon = True, framealpha=0.7, facecolor = 'white', loc='upper left') # Ajusta a cor, transparência e posição da Legenda
plt.grid(color='black', linewidth=.5) # Altera a cor e espessura da malha do gráfico
plt.axis((0,0.45,0,45)) # Ajusta os limites dos eixos
plt.tight_layout() # Melhora o enquadramento do gráfico 
#plt.show()
plt.savefig(dpi = 200, fname = 'pontos1.svg') # esta linha cria um arquivo svg do gráfico
```

```python
# Criando um intervalo para plotar a reta de ajuste
x1 = np.array(np.arange(0,0.50,0.05))
y_mod1 = a * x1 + b

plt.style.use('seaborn')
plt.figure(figsize = (7,5), dpi=180)
plt.plot(x, y, 'o', label = 'Pontos Experimentais', color = 'r')
plt.plot(x1, y_mod1, '-b', label = 'Modelo', linewidth = 2)
plt.xlabel(r'$\mathrm{I}$', family='sans-serif', weight = 'bold')
plt.ylabel(r'$\mathrm{V }$           ', rotation = 0,  family='sans-serif', weight = 'bold')
plt.legend(frameon = True, framealpha=0.7, facecolor = 'white', loc='upper left')
plt.grid(color='black', linewidth=.5)
plt.axis((0,0.45,0,45))
plt.tight_layout()
#plt.show()
plt.savefig(dpi = 200, fname = 'curva ajustada 1.svg') # esta linha cria um arquivo png com o gráfico
```
