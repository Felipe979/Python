# Carregando pacotes 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sympy import * # Para criar variáveis que plotam funções completas
```

# Inserção de Dados

```python
# Criando os vetores que são dados de entrada
x = np.array([0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40])
y = np.array([7.1, 9.6, 16.9, 21.0, 25.4, 28.1, 35.7, 39.0])
print('x = {}'.format(x))
print('=' * 40)
print('y = {}'.format(y))
print('=' * 40)

# Ajustando a curva e descobrindo os valores de a e b da equação y = ax + b

mod_linear = np.polyfit(x, y, 1) # A função polyfit retorna uma lista com os seguintes dados: [Coeficiente Angular, Coeficiente Linear]

a = float(mod_linear[0])
b = float(mod_linear[1])
print('a = {:.4f}'.format(a))
print('=' * 40)
print('b = {:.4f}'.format(b))
print('=' * 40)
print('y_mod = {:.4f}x + {:.4f}'.format(a,b))
print('=' * 40)

y_mod = a * x + b # Com os valores de 'a' e 'b', podemos construir uma função e calcular todos os valores teóricos fornecidos por ela
print('y_mod = {}'.format(y_mod))
print('=' * 40)

xr = np.arange(-0.1, 1, 1)

# Plotando os resultados
plt.style.use('ggplot')
plt.figure(figsize = (9,6), dpi=150)
plt.plot(x, y, 'o', color = 'g', label = 'Pontos Experimentais')
#plt.plot(x, y_mod, label = 'Modelo')
plt.plot(xr,  a * xr + b , linestyle = '-.', label = 'Teste', color = 'm')
plt.xlabel(r'$\mathrm{I\:(A)}$')
plt.ylabel(r'$\mathrm{V\:(V)}$          ', rotation = 0)
plt.title('Ajuste da Reta de Ohm', fontsize = 16, fontweight = 'bold', fontfamily = 'serif' )

plt.legend()
#plt.grid(color='black', linewidth=0.5)
plt.show()
```
