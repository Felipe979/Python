## Importando a biblioteca necessária

```python
import folium
from folium.plugins import MiniMap
```

### Criando um mapa básico

```python
popuptext = "<b>Monumento Marco Zero do Equador<\b>"  # texto utilizado na etiqueta de identificação local

# coordenadas geográficas em unidades decimais
latitude = 0.000833
longitude = -51.077778

# criação do mapa e definição de zoom
MarcoZero = folium.Map(location=[latitude, longitude], zoom_start=16)

MarcoZero
```

```python
popuptext = "<b>Monumento Marco Zero do Equador</b>"  # texto html utilizado na etiqueta de identificação local

latitude = 0.000833
longitude = -51.077778

MarcoZero = folium.Map(location=[latitude, longitude], zoom_start=16)

# inserindo um marcador em um local qualquer do mapa
folium.Marker(
    location=[0.000833,-51.077778], # localização do marcador
    popup=popuptext                 # nome para o marcador
    ).add_to(MarcoZero)             # adicionando as modificações no mapa

MarcoZero
```

```python
popuptext = "<b>Monumento Marco Zero do Equador<\b>"

latitude = 0.000833
longitude = -51.077778

MarcoZero = folium.Map(location=[latitude, longitude], zoom_start=16)

folium.Marker(location=[0.000833, -51.077778],
              popup=popuptext
             ).add_to(MarcoZero)

# inserindo um círculo de destaque em um local qualquer
folium.Circle(location=[0.000833, -51.077778],   # localização do círculo
              color="green",                     # cor da circunferência
              fill_color="orange",               # cor da área interna
              radius=50,                         # raio da circunferência, em metros
              weigth=4,
              fill_opacity=0.4,                  # opacidade da área interna
              tooltip="Marco Zero"               # nome da área do círculo
              ).add_to(MarcoZero)

MarcoZero
```

```python
popuptext = "<b>Monumento Marco Zero do Equador</b>"

latitude = 0.000833
longitude = -51.077778

MarcoZero = folium.Map(location=[latitude, longitude], zoom_start=16)

# alterando a estética do marcador
folium.Marker(location=[0.000833, -51.077778],
              popup=popuptext,
              icon=folium.Icon(color="orange",     # cor do marcador
                               icon="info-sign")  # ícone do marcador(pode ser uma nuvem, uma cruz, etc)
             ).add_to(MarcoZero)

folium.Circle(location=[0.000833, -51.077778],
              color="purple",
              fill_color="orange",
              radius=50,
              weigth=4,
              fill_opacity=0.4,
              tooltip="Marco Zero").add_to(MarcoZero)

MarcoZero
```
