# Importando as Bibliotecas Necessárias para plots bidimensionais

```python
import matplotlib.pyplot as plt # Biblioteca para criar e modificar gráficos
import numpy as np # Biblioteca para funções matemáticas
```

```python
x = [1,2,3,4,5]
y = [2,4,1,8,5]

plt.figure(dpi=150)
plt.scatter(x,y, color = "black")
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 5, 0.1) #Inserindo as coordenadas das abscissas
y = np.sin(x)                  #Inserindo uma função para as abscissas. Os valores das oordenadas também podem ser representados por pontos

plt.plot(x, y, color = 'magenta')
plt.show()
```

```python
x = np.arange(0, 5, 0.1) #Inserindo as coordenadas das abscissas
y = np.sin(x)                  #Inserindo uma função para as abscissas. Os valores das oordenadas também podem ser representados por pontos

plt.plot(x, y, "ro")
plt.show()
```

```python
import numpy as np
x1 = np.arange(-1000,1000,10) # variável criada = np.arange(valor inicial, valor final, incremento/decremento). A função np.arange cria uma lista que vai de um valor
                                                  # inicial até um valor final com um incremento ou decremento!

plt.plot(x1,  x1**3+4, label='Dados', color='k', lw='1.5') #, marker='v') #plt.plot(variável ind., variável dep., legenda, cor, espessura da curva, tipo de marcador da curva)
plt.xlabel('Abscissas', color='purple')                                                   #plt.xlabel(nome do eixo x, cor)
plt.ylabel('Ordenadas',color='blue')                                                      #plt.ylabel(nome do eixo y, cor)
plt.title('Teste de Gráfico')                                                                    #Título do gráfico
plt.xticks([-1000,-500,0,500,500])
#plt.yticks([-1,-0.5,0,0.5,1])
plt.legend()
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,1,8,5]

plt.plot(x,y, label='Dados', color='darkgreen', lw='1.5',marker='o')
plt.xlabel('Abscissas')
plt.ylabel('Ordenadas')
plt.title('Teste de Gráfico')
plt.legend()
plt.grid(color='deepskyblue')
plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')

# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Reds')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.grid()
plt.show()
```
